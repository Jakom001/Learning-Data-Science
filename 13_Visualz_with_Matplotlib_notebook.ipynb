# **DATA VISUALIZATION**

Using Matplotlib

import the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Load the Dataset

```python
df = pd.read_csv("https://raw.githubusercontent.com/dphi-official/Datasets/master/Standard_Metropolitan_Areas_Data-data.csv")
```

```python
# show the first 5 lines
df.head()
```

Visualize with matplotlib

1. Scatter plot

Use to represents values for two different numeric variables.

```python
plt.scatter(df.crime_rate, df.percent_senior)
plt.show
```

Adding Titles and labels

```python
plt.scatter(df.crime_rate, df.percent_senior)

plt.title('Plot of Crime Rate vs Percent Senior')
plt.xlabel("Crime Rate")
plt.ylabel("Percent Senior")
plt.show()
```

Line chart using matplotlib

Used to represent data over a continous life span.

```python
plt.plot(df.work_force, df.income)
plt.xlabel('Work Force')
plt.ylabel('Income')
```

Changing the Size of the plot

plt.figure(figsize=(new_width_pixels, new_height_pixels))

```python
plt.figure(figsize=(12,6)) # modify the size 12 x5 plot
plt.plot(df.work_force, df.income)
```

Style Formatting

```python
plt.plot(df.work_force, df.income, linestyle ='--', marker='o', color='r')
```

shortcut

```python
plt.plot(df.work_force, df.income, '--go') # green circles
```

Plotting Consucutive plots

```python
plt.plot(df.work_force, df.income, color='r')
plt.show()
plt.plot(df.physicians, df.income, color='b')
plt.show()
```

If you don't use plt.show() both variables will be plotted on the same figure.

Adding a legend

```python
plt.plot(df.work_force, df.income, color='r', label="work_force")
plt.plot(df.physicians, df.income, color='b', label = "phycisians")
plt.legend()
```

Multiple plots

.subplot() - used to add multiple plots in one figure.
It takes nrows, ncols, index

```python
plt.subplot(1,2,1)
plt.plot(df.work_force, df.income, 'go')
plt.title("income vs Work Force")

plt.subplot(1,2,2).label_outer()

plt.plot(df.hospital_beds, df.income, 'r^')
plt.title("Income vs Hospital Beds")

plt.suptitle('Sub Plots')
plt.show()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(6,6)) #creating a grid of 2 rows, 2 columns and 6x6 figure size
ax[0,0].plot(df.work_force, df.income, "go") # The top-left axes
ax[0,1].plot(df.work_force, df.income, "bo") # The top-right axes
ax[1,0].plot(df.work_force, df.income, "yo") # The bottom-left axes
ax[1,1].plot(df.work_force, df.income, "ro") # The bottom-right axes

plt.show()
```

```python
plt.hist(df.percent_senior)
plt.title("Histogram")
plt.xlabel("Percentage of senior Citizens")
plt.ylabel('Frequency')
```

Bar charts

Vertical Bar chart

```python
plt.bar(df.region, df.crime_rate, color='blue')
plt.title("Bar graph")
plt.xlabel('Region')
plt.ylabel('Crime Rate')
```

Horizontal Barchar

```python
plt.barh(df.region, df.crime_rate, color='blue')
plt.title("Bar graph")
plt.xlabel('Region')
plt.ylabel('Crime Rate')
```

Bar chart using multiple quantities.


```python
divisions = ["A", "B", "C", "D", "E"]
division_avg = [70, 82, 73, 65, 68]
boys_avg = [68, 67, 77, 61, 70]

# Using the NumPy arange function to generate values for index between 0-4.
# Here,stop is 5, start is 0, and step is 1
index = np.arange(5)
width = 0.30

plt.bar(index, division_avg, width, color="green", label="Division Marks")
plt.bar(index+width, boys_avg, width, color="blue", label="Boys Marks")

plt.title("Bar Graph")
plt.xlabel("Divisions")
plt.ylabel("Marks")
plt.legend()
plt.show()
```

Stacked Bar chart

```python
divisions = ["A", "B", "C", "D", "E"]
girls_avg = [72, 97, 69, 69, 66]
boys_avg = [68, 67, 77, 61, 70]

index = np.arange(5)
width = 0.50

plt.bar(index, boys_avg, width, color="green", label="Boys Marks")
plt.bar(index, girls_avg, width, color="blue", label="Girls Marks", bottom=boys_avg)

plt.title("Bar Graph")
plt.xlabel("Divisions")
plt.ylabel("Marks")
plt.legend()
plt.show()
```

Pie Chart

```python
firms = ["Firm A", "Firm B", "Firm C", "Firm D", "Firm E"]
market_share = [20,25,15,10,20]

# Explode the pie chart to emphasize a certain part or some parts( Firm B in this case)
# It is useful because it makes the highlighted portion more visible.
Explode = [0,0.1,0,0,0]

plt.pie(market_share, explode=Explode, labels=firms, autopct='%1.1f%%', startangle=45)

plt.axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.

plt.legend(title="List of Firms")

plt.show()
```
