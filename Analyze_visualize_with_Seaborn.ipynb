**Analyzing data and visualizing with Seaborn**


---



```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Loading the datasets
Let's use the inbuilt dataset provided by seaborn name tips

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
tips.describe()
```

**Univariate Plots**

These are based on a single variable and show the frequency of uniques values of a given variable


**Histogram**


```python
sns.distplot(tips['tip'], kde=False, bins=10)
```

Kerner Density Estimate <br> Is a way to estimate the probability desnsity function of a contious random variable

```python
sns.distplot(tips['tip'], hist=False, bins=10)
```

**Bivariate Plots**
<br> Used to find the relation between two variables and to find how the value of one variable changes the value of another variable

**Statistical Data Types**

Scatterplot

```python
sns.relplot(x="total_bill", y='tip', data=tips)
```

The third attribute can be used using the hue attribute in searborn

```python
sns.relplot(x="total_bill", y='tip', hue='smoker', data=tips)
```

Lineplot

```python
sns.relplot(x="total_bill", y="tip", kind="line", data=tips)
```

```python
sns.relplot(x="total_bill", y="tip", sort="False", kind="line", data=tips)
```

**Categorical Data types**

Scatterplot

```python
sns.catplot(x="sex", y="tip", data=tips)
```

```python
sns.catplot(x="day", y="total_bill", kind='swarm', hue='sex', data=tips)
```

```python
sns.catplot(x='smoker', y='tip', order=["No", "Yes"], data=tips)
```

Boxplot <br>
Shows the distribution of quantitative data in a way that facilitates comparison between variables or across levels of a categorical variable.

```python
sns.catplot(x="day", y="total_bill", kind="box", data=tips)
```

```python
sns.catplot(x="day", y="total_bill", hue="sex", kind="box", data=tips)
```

Violin Plot

This plot is used to visualize the distribution of the data and its probability density.

```python
sns.catplot(x='day', y="total_bill", hue='sex', kind="violin", data=tips)
```

Point Plots

A point plot represents an estimate of central tendency for a numeric variable by the position of scatter plot points and provides some indication of the uncertainty around that estimate using error bars.

```python
sns.catplot(x="day", y="total_bill", hue="sex", kind="point", data=tips)
```
